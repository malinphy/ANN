```python
url = 'https://raw.githubusercontent.com/malinphy/datasets/main/fifa/players_15.csv'
df = pd.read_csv(url, 
                 usecols = selected_col
                 ).dropna(axis=0)
```

```python
df_np = preprocessing.normalize(df.iloc[:,1:-1])
```

```python
df.iloc[:,1:-1]
```

```python
df.shape
```

```python
k = int(np.sqrt(df_np.shape[0]))
```

```python
searcher = scann.scann_ops_pybind.builder(df_np, 10, "dot_product").tree(
    num_leaves=k, num_leaves_to_search=int(k/20), training_sample_size=2500).score_brute_force(2).reorder(7).build()

# querying the searcher
```

```python
df[df.short_name == 'A. Robben']
```

```python
df[df.short_name == 'A. Robben']
```

```python
df_q = df[df.short_name == 'A. Robben']
df_q
```

```python
np_q = preprocessing.normalize(df_q.iloc[:,1:-1])
```

```python
neighbors, distances = searcher.search(np_q[0], final_num_neighbors=20)
```

```python
q_df = df.iloc[neighbors, :]
```

```python
q_df.index = q_df['short_name']
```

```python
q_df['distance'] = list(distances)
q_df['distance']
```

```python
sim_score = []
for x in df_np:
    sim_score.append(
        cosine_similarity(x.reshape(-1, 1), np_q.reshape(-1, 1)).mean())
    
```

```python
df['distance'] = sim_score
```

```python
df_sort = df.sort_values(by=['distance'], ascending=False)
```

```python
df_sort.index = df_sort['short_name']
```
